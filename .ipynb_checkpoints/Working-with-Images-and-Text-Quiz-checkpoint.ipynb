```python
text = ""
with open("gutenberg.txt", "r", encoding="utf-8") as f:
    text = f.read()
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
stop = set(stopwords.words("english"))
```

```python
remaining = re.split("\W+", text.lower())
print(len(remaining))
remaining = [w for w in remaining if w not in stop]
print(len(remaining))
```
